```python
from ultralytics import YOLO
import cv2
import argparse
import torch
import time
```

```python
def main():
    
    
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    
    # 모델 로드하기
    model = YOLO(model='yolov8m-pose.pt')

    while True:

        ret, frame = cap.read()
        result = model(source = frame, conf= 0.5, show=True , verbose = False) # save = True
        print(result[0].boxes.xyxy.cpu().numpy())
        cv2.imshow("HometestCAM", frame)
        
        # if(cv2.waitKey(0) == 27): # Askii table escape == 27
        #     break 
```

```python
main()
```
